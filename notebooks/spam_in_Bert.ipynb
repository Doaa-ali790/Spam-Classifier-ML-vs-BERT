```python
import pandas as pd
from transformers import pipeline

# 1. تحديد مسار الملف في جوجل درايف
# عادة ما يبدأ بـ /content/drive/MyDrive/
file_path = '/content/drive/MyDrive/smsspamcollection.tsv'

# 2. قراءة البيانات
try:
    df = pd.read_csv(file_path, sep='\t')
    print("✅ تم تحميل الملف بنجاح!")
except Exception as e:
    print(f"❌ لم يتم العثور على الملف. تأكد من المسار. الخطأ: {e}")

# 3. تشغيل خط إنتاج BERT (الموديل المتطور)
if 'df' in locals():
    print("جاري تحميل نموذج BERT المتطور...")
    classifier = pipeline("text-classification",
                          model="mrm8488/bert-tiny-finetuned-sms-spam-detection")

    # تطبيق التجربة على أول 10 رسائل من ملفك
    sample_messages = df['message'].head(10).tolist()
    results = classifier(sample_messages)

    print("\n--- نتائج تحليل BERT للرسائل في درايف ---")
    for i, res in enumerate(results):
        label = "SPAM 🚫" if res['label'] == 'LABEL_1' else "HAM ✅"
        print(f"الرسالة: {sample_messages[i][:50]}...")
        print(f"النتيجة: {label} (الدقة: {res['score']:.2%})\n")
```
